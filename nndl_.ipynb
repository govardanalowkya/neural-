```python
import numpy as np
# Create two {2,3} tensors with random values
A = np.random.rand(2, 3)
B = np.random.rand(2, 3)
# Element-wise addition
C = A + B
print("Tensor A:\n", A)
print("Tensor B:\n", B)
print("A + B:\n", C)
```

```python
import tensorflow as tf

def main():
    # Create two {2,3} tensors with random values
    tensor_a = tf.random.uniform(shape=(2, 3), minval=0.0, maxval=1.0)
    tensor_b = tf.random.uniform(shape=(2, 3), minval=0.0, maxval=1.0)

    # Element-wise addition
    result = tf.add(tensor_a, tensor_b)

    # Print tensors
    print("Tensor A:")
    print(tensor_a.numpy())

    print("\nTensor B:")
    print(tensor_b.numpy())

    print("\nElement-wise Addition (A + B):")
    print(result.numpy())

if __name__ == "__main__":
    main()
```

```python
import random

def main():
    # Create two {2,3} tensors (lists of lists) with random values
    tensor_a = [[random.random() for _ in range(3)] for _ in range(2)]
    tensor_b = [[random.random() for _ in range(3)] for _ in range(2)]

    # Element-wise addition
    result = [
        [tensor_a[i][j] + tensor_b[i][j] for j in range(3)]
        for i in range(2)
    ]

    # Print results
    print("Tensor A:")
    for row in tensor_a:
        print(row)

    print("\nTensor B:")
    for row in tensor_b:
        print(row)

    print("\nElement-wise Addition (A + B):")
    for row in result:
        print(row)

if __name__ == "__main__":
    main()
```
